```python
import requests
import time
positions=[]
for i in range(100):
    response= requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    data=response.json()
    latitude=data["latitude"]
    longitude=data["longitude"]
    time_stamp=data["timestamp"]
    positions.append((latitude, longitude, time_stamp))
    print("Point", i+1, ":", latitude, longitude)
    time.sleep(10)
```

```python
import matplotlib.pyplot as graph
latitudes=[]
longitudes= []
for position in positions:
    latitudes.append(position[0])
    longitudes.append(position[1])
graph.plot(longitudes, latitudes)
graph.xlabel("Longitude")
graph.ylabel("Latitude")
graph.show()
```
